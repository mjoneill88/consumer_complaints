```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
%matplotlib inline


c_c = pd.read_csv('complaints_dec_2014.csv')

c_c.Product.value_counts().plot(title='Product Types', kind = 'barh')

'Graph 1. shows a plot of the number of complaints organized by the type of product or service provided by the company.'
```

```python
c_c.Company.value_counts().head(10).plot(title = 'Complaints by Company Response', kind='barh')

'Graph 2 shows a plot of the number of complaints for the top 10 most-complained-about companies.'

```

```python
data = pd.read_csv('complaints_dec_2014.csv')
date_series = data.pop('Date received')
data.index = pd.to_datetime(date_series, format='%m/%d/%Y')
data['weekday'] = data.index.weekday
data = data.replace({'weekday': {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday',
                                   3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}})
data.weekday.value_counts()
data.weekday.value_counts().plot(title='Number of Complaints by Day of Week', kind='barh')
plt.show()

'Graph 4 shows the mean number of complaints by day of the week'
```
